```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
webpage = requests.get("https://en.wikipedia.org/wiki/List_of_solar_storms").content
```

```python
soup = BeautifulSoup(webpage,"html.parser")
```

```python
dates_list = soup.find_all('td')
```

```python
len(dates_list)
```

```python
# dates = []
# dates_list = soup.find_all('td')
# for i in list(range(0,105,3)):
#     dates.append(date_list[i].get_text())
dates_list = soup.find_all('td')
dates_index = list(range(0,105,3))
a = []
only_dates = [dates_list[i] for i in dates_index]
for i in only_dates:
    a.append(i.get_text())
```

```python
dates = a[:-3]
dates_final = [i.strip('\n') for i in dates]
```

```python
event_index = list(range(1,105,3))
only_events = [dates_list[i] for i in event_index]
```

```python
events = only_events[:-3]
events_final = [i.get_text().strip('\n') for i in events]
```

```python
significance_index = list(range(2,105,3))
only_significance = [dates_list[i] for i in significance_index]
```

```python
significance = only_significance[:-3]
significance_final = [i.get_text().strip('\n') for i in significance]
```

```python
d = {'Date':dates_final, 'Event':events_final , 'Significance':significance_final}
```

```python
df = pd.DataFrame.from_dict(d)
```
